```python
from pathlib import Path
from fast_elmo.load_data.batch_generator import BatchGenerator
```

**BatchGenerator** -- это класс, генерирующий батчи для обучения языковой модели. 

Для работы ему нужно передать путь до папки, где лежат файлы *train.txt, dev.txt, test.txt* (в виде объекта Path), плюс параметры батчей (вообще, параметров должно быть больше, позже я настрою их проброс до нужных классов. Пока что все остальное по умолчанию).

При инициализации ему потребуется построить словарь "слово-индекс". Этот процесс занимает некоторое время (что тоже можно и нужно оптимизировать, добавив подгрузку уже сохраненного словаря)

```python
path_to_data = Path('../data/Lenta_ru/')
batch_generator = BatchGenerator(path_to_data=path_to_data, batch_size=4)
```

Батч представляет из себя словарь со следующими полями:
* **raw_text** -- текст, который дальше будет отправляться в эмбеддер от allennlp
* **forward_target** -- тензор из целевых переменных (индексы слов) для forward-подсети
* **backward_target** -- тензор из целевых переменных (индексы слов) для backward-подсети

Батчи получаются командой:

```python
batch_generator.generate_train_batch()
```

Время загрузки одного батча:

```python
%timeit batch_generator.generate_train_batch()
```
